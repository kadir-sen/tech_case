```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.grid'] = True
sns.set_style('whitegrid')

from src.data.loader import load_validated

import json
from pathlib import Path
```

> Eğitim uzun sürer. Headless: `python -m src.models.train`. Sonuçlar `artifacts/eval/all_models.json`'a yazılır.

```python
eval_path = Path('../artifacts/eval/all_models.json')
if not eval_path.exists():
    print('Henüz eğitilmemiş. Önce `python -m src.models.train` çalıştırın.')
else:
    res = json.loads(eval_path.read_text())
    print('models trained:', list(res['models'].keys()))
    print('split cutoffs:', res['split'])
```

## Test setinde PR-AUC karşılaştırması

```python
if eval_path.exists():
    rows = []
    for name, m in res['models'].items():
        rows.append({'model': name, 'pr_auc_test': m['test']['core']['pr_auc'], 'roc_auc_test': m['test']['core']['roc_auc']})
    cmp = pd.DataFrame(rows).sort_values('pr_auc_test', ascending=False)
    print(cmp.to_string(index=False))
```
